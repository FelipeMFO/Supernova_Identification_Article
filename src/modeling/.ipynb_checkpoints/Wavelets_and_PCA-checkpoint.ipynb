```python
def wavelets(obj, wavelet = 'sym2', mlev = 2 ):
    wav = wt.Wavelet(wavelet)
    coeff_g = np.array(wt.swt(obj['desg_GP'][0], wav, level=mlev)).flatten()
    coeff_i = np.array(wt.swt(obj['desi_GP'][0], wav, level=mlev)).flatten()
    coeff_r = np.array(wt.swt(obj['desr_GP'][0], wav, level=mlev)).flatten()
    coeff_z = np.array(wt.swt(obj['desz_GP'][0], wav, level=mlev)).flatten()
    coeffs = np.concatenate([coeff_g,coeff_i,coeff_r,coeff_z])
    #coeffs = coeffs.reshape(-1, 1)
    return coeffs
```

```python
df_GP_wav = pd.read_pickle("../../data/processed/df_GP_wav.pkl")
```

```python
df_GP_wav
```

```python
s = df_GP_wav['wavelets']
rslt = np.array(s.tolist())
print(type(rslt))
print(rslt.shape)

pca = PCA(n_components=20) #CHANGE HERE THE NUMBER OF COMPONENTS YOU WANT
pca.fit(rslt)  
print(pca.explained_variance_ratio_)
print(pca.singular_values_)
features = pca.fit_transform(rslt)
print(features.shape)

df_features = pd.DataFrame(features, columns=['f1', 'f2','f3', 'f4','f5',
                                              'f6', 'f7','f8', 'f9','f10',
                                              'f11', 'f12','f13', 'f14','f15',
                                              'f16', 'f17','f18', 'f19','f20'])
df_features
```

```python
df_GP_wav_PCA20 = pd.concat([df_GP_wav, df_features], axis = 1)
df_GP_wav_PCA20
```

```python
df_GP_wav_PCA20.to_pickle("./df_GP_wav_PCA20.pkl")
```
